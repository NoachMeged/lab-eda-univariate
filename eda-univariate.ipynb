```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate IQR
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1

# Determine the upper bound for outliers
upper_bound = q3 + 1.5 * iqr

# Calculate the 95th percentile
percentile_95 = df['price'].quantile(0.95)

# Print thresholds
print(f"Upper Bound for Outliers: £{upper_bound:.2f}")
print(f"95th Percentile: £{percentile_95:.2f}")

# Create the box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='price', color='skyblue')

# Set x-axis limit to focus on typical prices
plt.xlim(0, percentile_95 * 2)  # Adjust to show outliers clearly
plt.title('Box Plot of Product Prices with Outliers')
plt.xlabel('Price (£)')
plt.show()
```

```python
# How do customers generally rate products? Is there a common trend? Most people tend to give o zero rating. Costumers tend to give a 4 star rating after or 3 before a 5 star rating



import pandas as pd
import matplotlib.pyplot as plt



# Assuming you already have your DataFrame 'df'
stars = df['stars']

# Plotting the histogram to show the spread of ratings
plt.figure(figsize=(8, 6))
plt.hist(stars, bins=range(int(stars.min()), int(stars.max()) + 2), edgecolor='black', color='skyblue', alpha=0.7)

# Adding labels and title
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating (Stars)")
plt.ylabel("Frequency")
plt.xticks(range(int(stars.min()), int(stars.max()) + 1))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Determine the variance, standard deviation, and interquartile range for product rating.

# Assuming you already have your DataFrame 'df'
stars = df['stars']

# Calculate variance
variance = stars.var()

# Calculate standard deviation
std_deviation = stars.std()

# Calculate the interquartile range (IQR)
iqr = stars.quantile(0.75) - stars.quantile(0.25)

# Print the results
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")
print(f"Interquartile Range (IQR): {iqr}")
```
